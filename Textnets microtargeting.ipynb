```python
from textnets import Corpus, Textnet
import pandas as pd
import igraph

data = pd.read_csv('./microtargeting_microtarget_oldtweets2.csv',
                   names = ['username','text','date'])

corpus = Corpus(pd.Series(data.text.unique())) # The "unique()" eliminates duplicate tweets
np = corpus.noun_phrases()

tn = Textnet(np, min_docs=10)

words = tn.project(node_type='term')
words.plot(label_nodes=True,
           show_clusters=True)
```
